# Programming Question 1

Compute and display the mean of the faces. Then subtract the
mean from all the faces. You are expected to display faces instead
of vectors.

```python
df = pd.read_csv('Yale.csv', header=None)
label = df[0]
img = df.drop(df.columns[0], axis = 1)
```

```python
import scipy.misc
from scipy import ndimage
import matplotlib.pyplot as plt
#example image before subtracting mean
first_image = img.iloc[2]
first_image = np.array(first_image, dtype='float')
pixels = first_image.reshape((32, 32))
plt.imshow(pixels, cmap='gray')
plt.show()
```

```python
#finding mean
df2 = img.mean(axis = 0)
mean_image = df2
#display mean
mean_image = np.array(mean_image, dtype='float')
pixels = mean_image.reshape((32, 32))
plt.imshow(pixels, cmap='gray')
plt.show()
```

```python
#subtrcacting mean
aft_mean_img = img - mean_image
#example image after subtracting mean
temp = aft_mean_img.iloc[2]
temp = np.array(temp, dtype='float')
pixels = temp.reshape((32, 32))
plt.imshow(pixels, cmap='gray')
plt.show()
```

(b) Perform PCA on the whole data set. This can be down through
the singular value decomposition of the data matrix.
1. Display the singular values;
2. Display the top 10 principal components. You are expected
to display faces instead of vectors.

```python
svd_img = np.array(img.iloc[0]).reshape((32,32))
print(svd_img)
```

```python
from sklearn.decomposition import PCA
pca = PCA().fit(img)
print(pca.singular_values_)
```

```python
Face = pca.components_[:10]
for i in range(10): 
    print(Face[i])
    temp = Face[i]
    temp = np.array(temp, dtype='float')
    pixels = temp.reshape((32, 32))
    plt.imshow(pixels, cmap='gray')
    plt.show()
```

(c) Take an arbitrary face, display the reconstructed faces using k
principal components while k = 10,50,100,500. Plot the recon-
struction error as k increases. You can try k = 1,50,100,150,200,....
Denote the original and reconstructed images by x and xkrespec-
tively. Then the reconstruction error can be defined as xxk.

```python
k = [10, 50, 100,150]
for i in k:
    Face = pca.components_[:i+1]
    print(Face[i])
    temp = (mean_image - Face[i])**2
    temp = np.array(temp, dtype='float')
    pixels = temp.reshape((32, 32))
    plt.imshow(pixels, cmap='gray')
    plt.show()
```

# Programming Question 2

generate the dataset

```python
import math
import random
import seaborn as sns
import pandas as pd
data_1 = [] 
data_2 = []
for i in range(200): 
    data_1.append(random.uniform(0, 2*math.pi))
    data_2.append(random.uniform(0, 2*math.pi))
x1_1 = []
x2_1 = []
x1_2 = []
x2_2 = []
for i in range(len(data_1)):
    x1_1.append(1*math.cos(data_1[i]))
    x2_1.append(1*math.sin(data_1[i]))
    x1_2.append(2*math.cos(data_2[i]))
    x2_2.append(2*math.sin(data_2[i]))
x1 = x1_1+x1_2
x2 = x2_1+x2_2
df = pd.DataFrame(list(zip(x1, x2)),columns =['x1', 'x2'])
noise = np.random.normal(0, 0.05, [400,2])
df1 = df + noise
#plt.plot(x1, x2, 'o', )
sns.scatterplot(data = df1,x='x1',y='x2')
```

(a) Apply K means with K = 2, and display the clustering results.

```python
from sklearn.cluster import KMeans
kmean = KMeans(n_clusters=2, random_state=3425).fit(df1)
kmeanresult = kmean.predict(df1)
sns.scatterplot(data = df1,x='x1',y='x2', hue=kmean.predict(df1))
```
